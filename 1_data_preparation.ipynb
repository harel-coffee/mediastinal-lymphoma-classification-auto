```python
import os

data_dir  = "./data"
data_file = "Database_Emato_mediastinalbulky_anonimized.csv"
file_path = os.path.join ( data_dir, data_file )
```

```python
import pandas as pd

db = pd . read_csv ( file_path, sep = ";", header = 0 )
db
```

## Data cleaning

- empty rows at the end of the database removed
- columns with no clear contents dropped
  - `PARAMS_IntensityResampling`
  - `PARAMS_BoundsRangeOfValueAfterDiscretisation(SUVbw)`
  - `CHECK_Cluster(s)ToSmall`
- data format for dates and float number fixed
- age information recovered

```python
db . dropna ( subset = ["ID"], inplace = True )   # remove empty rows
db
```

```python
### Data content not clear
cols_to_drop = [
    "PARAMS_IntensityResampling", "PARAMS_BoundsRangeOfValueAfterDiscretisation(SUVbw)", "CHECK_Cluster(s)ToSmall"
               ]

db . drop ( labels = cols_to_drop, axis = 1, inplace = True )
db
```

```python
from utils import data_cleaning

### Fix date format
for cname in [ "Data nascita", "dPET staging" ]:
  data_cleaning ( db, col_name = cname, target_type = "date", inplace = True )

### Fix float number format
cols = list ( db.columns [2:] )
for cname in cols:
  data_cleaning ( db, col_name = cname, target_type = "float", inplace = True )

db
```

```python
age = list()

for i in range ( len(db) ):
  age . append ( db["dPET staging"][i].year - db["Data nascita"][i].year )

db["age"] = age   # add age information

db . drop ( labels = ["Data nascita", "dPET staging"], axis = 1, inplace = True )
db
```

## Data renaming

- label column renamed to `lymphoma_type`
  - `lymphoma_type = 1.0` --> HL
  - `lymphoma_type = 2.0` --> GZ
  - `lymphoma_type = 3.0` --> PML
- `age` feature moved to the third column

```python
cols = list ( db.columns )
cols[1] = "lymphoma_type"
db . columns = cols

cols . remove ( "age" )
cols = cols[:2] + ["age"] + cols[2:]

db = db[cols]
db
```

## Data export

```python
exp_data_file = "db_mediastinalbulky_v1-cleaned.pkl"
exp_file_path = os.path.join ( data_dir, exp_data_file )

db . to_pickle ( exp_file_path )
```
